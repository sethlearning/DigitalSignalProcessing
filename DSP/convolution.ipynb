```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def square(x):
    return 0 if x < 3 or x > 5 else 2

def triangle(x):
    return 0 if x < 0 or x > 2 else x
```

```python
signal_a = [square(x/100) for x in range(1000)]
signal_b = [triangle(x/100) for x in range(202)]
```

```python
plt.plot([x for x in range(1000)], signal_a)
plt.plot([x for x in range(202)], signal_b)
```

```python
conv = (len(signal_a) - len(signal_b)) * [0]
len(conv)
```

```python
for l in range(len(conv)):
    for i in range(len(signal_b)):
        conv[l] += signal_a[l + len(signal_b) - i] * signal_b[i]
    # conv[l] /= len(signal_b)

plt.plot(conv)
plt.show()
```

```python
conv2 = np.convolve(signal_a, signal_b, mode='valid')
# conv2 /= len(signal_b)
plt.plot(conv2)
plt.show()
```
