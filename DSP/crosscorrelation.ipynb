```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal_a = np.sin(np.r_[-1:10:.1])
signal_b = np.sin(np.r_[-1:5:.1] + np.pi/4)
# np.r_[-1:1:50j, [0]*3, 5, 6]
# print(f'{(np.r_[-1:1:.1]):.2f}')
print(len(signal_a), len(signal_b))
```

```python
plt.plot(signal_a)
plt.plot(signal_b)
```

```python
corr = (len(signal_a) - len(signal_b)) * [0]

for l in range(len(corr)):
    corr[l] = sum(signal_a[i+l] * signal_b[i] for i in range(len(signal_b)))
```

```python
plt.plot(corr)
plt.show()
```

```python
corr2 = np.correlate(a=signal_a, v=signal_b)
plt.plot(corr2)
plt.show()
```
