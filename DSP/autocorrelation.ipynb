```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [3, 16, 156, 47, 246, 176, 233, 140, 130, 
        101, 166, 201, 200, 116, 118, 247, 
        209, 52, 153, 232, 128, 27, 192, 168, 208, 
        187, 228, 86, 30, 151, 18, 254, 
        76, 112, 67, 244, 179, 150, 89, 49, 83, 147, 90, 
        33, 6, 158, 80, 35, 186, 127]

lags = range(10)
```

```python
acorr = len(lags) * [0]

mean = sum(data) / len(data)
variance = sum([(x - mean)**2 for x in data]) / len(data)

print(mean, variance)
```

```python
# Normalized data
ndata = [x - mean for x in data]
print(ndata)
```

```python
for l in lags:
    c = 1 # Self-correlation
    if (l > 0):
        tmp = [ndata[l:][i] * ndata[:-l][i] for i in range(len(data) - l)]
        c = sum(tmp) / len(data) / variance
    acorr[l] = c

plt.plot(acorr)
plt.show()
```

```python
mean2 = np.mean(data)
variance2 = np.var(data)
ndata2 = data - mean2

acorr2 = np.correlate(ndata2, ndata2, mode='full')[len(ndata2)-1:]
acorr2 = acorr2 / variance2 / len(ndata)
plt.plot(acorr2)
plt.show()
```
