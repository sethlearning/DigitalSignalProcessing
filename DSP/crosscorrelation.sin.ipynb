```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sinx = np.sin(np.r_[0:100:.01])
cosx = np.cos(np.r_[0:100:.01])
print(len(sinx))
```

```python
plt.plot(sinx)
plt.plot(cosx)
plt.show()
```

```python
corr = len(sinx) * [0]

for l in range(len(corr)):
    sum = 0
    crosslength = len(sinx)-l
    for i in range(crosslength):
        sum += sinx[i+l] * cosx[i]
    # sum = sum / crosslength
    sum = sum / crosslength * 2
    corr[l] = sum

plt.plot(corr)
plt.show()
```

```python
corr2 = np.correlate(a=sinx, v=cosx, mode='same')
plt.plot(corr2)
plt.show()
```
