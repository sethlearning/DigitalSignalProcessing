```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sinx = np.sin(np.r_[0:100:.01])
cosx = np.cos(np.r_[0:100:.01])
print(len(sinx))
```

```python
plt.plot(sinx)
plt.plot(cosx)
```

```python
corrzero = 0
for i in range(len(sinx)):
    corrzero += sinx[i] * cosx[i]

print(corrzero)
print(corrzero/len(sinx))
```

```python
plt.plot([sinx[i]*cosx[i] for i in range(len(sinx))])
```

```python
corrzero = 0
for i in range(len(sinx)):
    corrzero += sinx[i] * sinx[i]

print(corrzero)
print(corrzero/len(sinx))
```

```python
plt.plot([sinx[i]*sinx[i] for i in range(len(sinx))])
```
